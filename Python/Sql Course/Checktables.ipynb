```python
import pandas as pd
import numpy as np

import sqlite3

```

```python
database = 'database.sqlite'

conn = sqlite3.connect(database)
print('Opened data successfully')
```

```python
df = pd.read_sql("""SELECT *
                 FROM sqlite_master
                 WHERE type='table';""", conn)
df
```
