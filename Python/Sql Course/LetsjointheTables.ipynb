```python
import numpy as np
import pandas as pd
import sqlite3
```

```python
database = 'database.sqlite'

conn = sqlite3.connect(database)

```

```python
tables = pd.read_sql("""SELECT *
                        FROM sqlite_master
                        WHERE type='table'""", conn)
tables
```

```python
joined_city = pd.read_sql("""SELECT *
                          FROM country c
                          INNER JOIN city ci
                          ON c.Country_id == ci.Country_id""", conn)
joined_city
```

```python
joined_left = pd.read_sql("""SELECT *
                          FROM player
                          LEFT JOIN season
                          ON player.Player_Id == season.Man_of_the_Series""", conn)

joined_left
```

```python
joined_cross = pd.read_sql("""SELECT c.Country_Id, c.Country_Name, ci.City_Name
                           FROM country c
                           CROSS JOIN city ci""", conn)
joined_cross

```

```python
union = pd.read_sql("""SELECT Player_Name
                    FROM player
                    UNION
                    SELECT Team_Name
                    FROM team""", conn)
union
```
